```python
# pipeline code
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, balanced_accuracy_score, roc_auc_score, f1_score, precision_score, recall_score
import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')

# load the pdf
@task(name='Task 1: Load PDF')
def load_pdf_tabula(file):
    '''
    Load PDF using tabula
    '''
    with pdfplumber.open(file) as pdf:
        # every page as an individual dataframe - list
        dfs = [page.extract_table() for page in pdf.pages]
        # concatenate
        df_pdf = pd.concat([pd.DataFrame(df) for df in dfs], ignore_index=True)
        df_pdf.columns = df_pdf.iloc[0]
        df_pdf = df_pdf.drop(0).reset_index(drop=True)
    print("Successfully loaded the pdf document!")
    print("It has {} number of rows".format(len(df_pdf)))
    df_pdf.info()
    return df_pdf

# load the excel
@task(name='Task 2: Load Excel File')
def load_excel(file):
    df_excel = pd.read_excel(file)
    print("Successfully loaded the Excel file!")
    return df_excel

# combine the two files
@task(name='Task 3: Merge the two files')
def combine_files(df_pdf, df_excel):
    '''
    horizontal concatination - both files have 21 cols
    '''
    df_new = pd.concat([df_pdf, df_excel], ignore_index=True)
    return df_new

# visualise the data
@task(name='Task 4: Visual Inspection - Bar Chart')
def visualize(df):
    '''
    Visualize to get some insights
    '''
    fig, ax = plt.subplots()
    (df.groupby('Churn')['Contract']
     .value_counts(normalize=True)
     .unstack() # turns the previous steps into a dataframe
    ).plot(kind='bar', ax=ax)
    ax.set_yticklabels(['{:,.0%}'.format(x) for x in ax.get_yticks()])
    ax.set(ylabel ='Proportion %', title='Churn by Contract Type')
    plt.xticks(rotation=45, ha='right');

@task(name='Task 5: Visual Inspection - Histogram')
def histogram(df):
    df['MonthlyCharges'] = df['MonthlyCharges'].astype(float)
    fig, ax = plt.subplots()
    ax= sns.kdeplot(df, x='MonthlyCharges', hue='Churn',
                   common_norm=False, ec='black', shade=True,
                alpha=0.6, lw=2, zorder=2);

@task(name='Task 6: Load to CSV')
def save_data(df, filename):
    df.to_csv(filename, index=False)
    print('Successfully loaded the dataframe to CSV!')

@flow
def etl_pipeline_tabula():
    # Step 1: load pdf
    pdf_file = "bpp_churn_data.pdf"
    df_pdf = load_pdf_tabula(pdf_file)
    # Step 2: load excel
    excel_file = 'bpp_churn_data.xlsx'
    df_excel = load_excel(excel_file)
    # Step 3: combine the files
    df_new = combine_files(df_pdf, df_excel)
    # Step 4: visualize
    visualize(df_new)
    histogram(df_new)
    # Step 5: save the final output
    new_file = "final_concat.csv"
    save_data(df_new, new_file)


etl_pipeline_tabula()
```
